```python
"""
Clean methylation data by removing NaN values.

"""
 
import dask.dataframe as dd
 
# 1. Load the raw data (Dask handles large CSVs efficiently)
data = dd.read_csv(r'data')  # <- replace 'data' with your actual file path
 
# 2. Convert to pandas for row-wise NaN dropping
df = data.compute()
 
print(f"Shape before cleaning: {df.shape}")
print(f"Total NaN values: {df.isna().sum().sum()}")
 
# 3. Drop rows containing any NaN values (matches the notebook's approach)
df_clean = df.dropna()
 
print(f"Shape after cleaning: {df_clean.shape}")
print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")
 
# 4. Save the cleaned data
df_clean.to_csv('data_clean.csv', index=False)
print("Cleaned data saved to data_clean.csv")
```
