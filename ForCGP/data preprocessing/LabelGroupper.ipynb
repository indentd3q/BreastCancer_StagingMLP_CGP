#### This notebook shows a placeholder example of how labeled data can be created; it is illustrative rather than a fully functional implementation.

```python
import dask.dataframe as dd
import pandas as pd
```

```python
data = dd.read_csv(r'data')
```

```python
data.head()
```

```python
pandas_df = data.compute()
```

```python
BIA = pd.read_csv('D:\Kuliah\Semester 6\Dataset\TCGA-BRCA.clinical.csv')
```

```python
pandas_df = pandas_df.dropna()
```

```python
yang_kepake2 = BIA[["sample", "ajcc_pathologic_stage.diagnoses"]]
yang_kepake2
```

```python
dask_df = pandas_df.set_index('Composite Element REF')
```

```python
Transposed =  dask_df.T
Transposed
```

```python
mapping = yang_kepake2.set_index('sample')['ajcc_pathologic_stage.diagnoses'].str.replace('Stage ', '')

UMN_Horizontal = Transposed.assign(Label=Transposed.index.map(mapping))
```

```python
Labeled = UMN_Horizontal
```

```python
Labeled['Label'].value_counts()
```

```python
Labeled.head()
```

```python
Labeled['Label'] = Labeled['Label'].apply(lambda x: 'Early' if x in ['I', 'IB', 'IV', 'IA', 'IIA', 'IIB', 'II', 'X'] else 'Late')
```

```python
Labeled['Label'].value_counts()
```

```python
Labeled.head()
```

```python
Labeled.to_csv('Labelled.csv', index=False)
```
