```python
import pandas as pd

phenotype = pd.read_csv(r"D:\Kuliah\Semester 6\Dataset\TCGA-BRCA.clinical.csv") ## This data can be downloaded from https://gdc-hub.s3.us-east-1.amazonaws.com/download/TCGA-BRCA.methylation450.tsv.gz

mirna = pd.read_csv(r"D:\Kuliah\Semester 6\Dataset\TCGA-BRCA.methylation450.csv") ## This data can be downloaded from https://gdc-hub.s3.us-east-1.amazonaws.com/download/TCGA-BRCA.clinical.tsv.gz

mirna_data = mirna

white_samples = phenotype[phenotype["race.demographic"].str.lower() == "white"]["sample"]

common_samples = set(white_samples) & set(mirna_data.columns)

mirna_white = mirna_data[["Composite Element REF"] + list(common_samples)]

mirna_white.to_excel("miRNA_white_races.xlsx", index=False)
```
