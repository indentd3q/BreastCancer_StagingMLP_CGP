# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import methylcheck, methylprep, methylize
```

# Data Import and Preprocess

```python
meth_data = pd.read_csv('RacialData/white_cg_data_dropped_self-labelled.csv', index_col=0, low_memory=False, sep=',')
meth_data.shape
```

```python
meth_data = meth_data.fillna(0)
meth_data['stage'].value_counts()
```

```python
beta_val = meth_data.iloc[:,:-1]
beta_val = beta_val.clip(lower=1e-6, upper=1 - 1e-6)
beta_val = beta_val.T
beta_val.shape
```

## Initialize methylcheck's sketchy probes list from the Epic array

```python
sketchy_probes_list = methylcheck.list_problem_probes(array='epic')
```

## Exclude CpG sites and get filtered dataframe

```python
excluded_df = methylcheck.exclude_probes(beta_val, sketchy_probes_list)
print(excluded_df.shape)
excluded_df.head()
```

```python
filtered_df = methylcheck.exclude_sex_control_probes(excluded_df, 'epic', no_sex=True, no_control=True, verbose=True)
```

```python
phenotypes = meth_data['stage'].tolist()
filtered_test_results = methylize.diff_meth_pos(filtered_df, phenotypes)
```

```python
filtered_test_results.to_csv("MethylizeResults/BRCA_filtered_28k_results.csv")
```
